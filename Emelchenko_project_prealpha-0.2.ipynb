```python
#2.1. Сетнимент анализ по plot файла wiki_movie_plots_deduped.csv на английском bert 
```

```python
SentimentClassifier = pipeline("sentiment-analysis")
```

```python

def FunctionBERTSentiment(inpText):
  return(SentimentClassifier(inpText)[0]['label'])

```

```python
movies['PlotNorm'] = movies.Plot.str[:512] #нормализация, чтобы смогла работать модель
```

```python
movies
```

```python
%%time
movies['Sentiment']=movies['PlotNorm'].apply(FunctionBERTSentiment)
```

```python
movies
```

```python
import matplotlib.pyplot as plt
fig, subPlot =plt.subplots(nrows=1, ncols=2, figsize=(10,4))
fig.suptitle("Sentiment analysis of wiki_movie_plots")
 
# Grouping the data
GroupedData=movies.groupby('Sentiment').size()
 
# Creating the charts
GroupedData.plot(kind='bar', ax=subPlot[0], color=['crimson', 'lightblue'])
GroupedData.plot(kind='pie', ax=subPlot[1], colors=['crimson', 'lightblue'])
```

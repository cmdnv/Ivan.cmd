```python
import requests
```

```python
offset_ = 7001
limit_ = 4000
url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
res = requests.get(url)
objects_data = res.json()
objects_data.get('data').get('list')[0]
```

```python
objects_list = objects_data.get('data').get('list')
objids = [x.get('objId') for x in objects_list]
objids[:10]
```

```python
url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/29679'
```

```python
res = requests.get(url)
d = res.json()
```

```python
import pandas as pd
from tqdm.auto import tqdm
```

```python
d = []
for obj in tqdm(objids):
        url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{obj}'
        res = requests.get(url)
        d.append(res.json())
```

```python
len(d)
```

```python
df1 = pd.json_normalize(d)
df1
```

```python
df3 = pd.json_normalize(d)
df3
```

```python
df2 = pd.json_normalize(d)
df2
```

```python
frames = [df1, df2, df3]
dff = pd.concat(frames)
dff.reset_index(drop=True, inplace=True)
dff
```

```python
writer = pd.ExcelWriter('output.xlsx')
dff.to_excel(writer)
writer.save()
```
